```python
import pandas as pd
import numpy as np
Company = pd.read_csv("Company_Data.csv")
```

```python
Company.iloc[:,0:8]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.barplot(x='Income', y='ShelveLoc',data=Company)
```

```python
Company
```

```python
# LabelEncoder
from  sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
(Company["ShelveLoc"]) = LE.fit_transform(Company['ShelveLoc'])
```

```python
def norm_fun(i):
  x = (i-i.min())/(i.max()-i.min())
  return(x)
```

```python
Company.hist()
```

```python
columns_to_plot = ['Sales', 'CompPrice', 'Income', 'Advertising', 'Population', 'Price', 'ShelveLoc', 'Age', 'Education']
for column_name in columns_to_plot:
    plt.figure(figsize=(10, 5))
    plt.plot(Company[column_name], marker='o', linestyle='-')
    plt.title(f'Line Plot for {column_name}')
    plt.xlabel('columns_to_plot')
    plt.ylabel('Income')
    plt.grid(True)
plt.show()
```

```python
# Normalized the DataFrame
df_norm = norm_fun(Company.iloc[:,0:6])
df_norm
```

```python
Company1 = Company.iloc[:,6]
Company1
```

```python
Company2 = pd.concat([df_norm,Company1],axis = 1)
Company2
```

```python
X1 = Company2.iloc[:,0:6]
Y1 = Company2.iloc[:,6]
X1
```

```python
Y1
```

```python
import seaborn as sns
sns.pairplot(data = Company2, hue = 'ShelveLoc')
```

```python
# Splitting the Data
from sklearn.model_selection import train_test_split
X1train, X1test, Y1train, Y1test = train_test_split(X1,Y1, test_size = 0.2, random_state = 0)
```

```python
print(X1train.shape)
print(X1test.shape)
print(Y1train.shape)
print(Y1test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
DTC = DecisionTreeClassifier(criterion='entropy', max_depth=3)
DTC.fit(X1train, Y1train)
```

```python
preds = DTC.predict(X1test)
```

```python
preds
```

```python
np.mean(preds==Y1test)*100
```

```python
from sklearn.metrics import classification_report
print(classification_report(preds,Y1test))
```

```python
from sklearn.model_selection import KFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
KFold = KFold(n_splits=10)
model = RandomForestClassifier(n_estimators=100,max_features=3)
results = cross_val_score(DTC,X1,Y1,cv = KFold)
print(results.mean()*100)
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.model_selection import KFold
Kfold = KFold(n_splits=10)
DTC1 = DecisionTreeClassifier()
modelBag = BaggingClassifier(base_estimator=DTC1, n_estimators=100, random_state=7)
results = cross_val_score(modelBag,X1,Y1,cv = Kfold)
print(results.mean()*100)
```
