```python
books_new = books.pivot_table(index='UserId', columns='Title', values='Rating')
```

```python
books_new
```

```python
# Replacing NaN to 0
books_new.fillna(0, inplace=True)
```

```python
books_new.index = books['UserId'].unique()
books_new
```

```python
from sklearn.metrics import pairwise_distances
person1 = 1 - pairwise_distances(books_new.values, metric='cosine')
person1
```

```python
np.fill_diagonal(person1, 0)
```

```python
# Saving person1 results in DataFrame
person = pd.DataFrame(person1)
person
```

```python
# Setting index and columns names to user id
person.index = books.UserId.unique()
person.columns = books.UserId.unique()
```
