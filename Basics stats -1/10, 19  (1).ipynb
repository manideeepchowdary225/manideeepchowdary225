```python
# Sample data for chick weights and the related variable
np.random.seed(42)
chick_weights = np.random.randint(100, 400, 100)
related_variable = np.random.uniform(0, 1, 100)

# Create a DataFrame
data = pd.DataFrame({'Chick Weight': chick_weights, 'Related Variable': related_variable})
```

```python
# Create a histogram for chick weights
plt.figure(figsize=(8, 6))
sns.histplot(data['Chick Weight'], bins=20, color='skyblue')
plt.xlabel('Chick Weight')
plt.ylabel('Frequency')
plt.title('Histogram of Chick Weights')

# Create a scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Related Variable', y='Chick Weight', data=data, color='green')
plt.xlabel('Related Variable')
plt.ylabel('Chick Weight')
plt.title('Scatter Plot of Related Variable vs. Chick Weight')

# Show the plots
plt.show()

```
