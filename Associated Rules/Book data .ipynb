```python
import pandas as pd
import numpy as np
Book = pd.read_csv("book.csv")
```

```python
Book.info
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
Book = apriori(Book, min_support=0.1, use_colnames=True)
Book
```

```python
rules = association_rules(Book, metric="lift", min_threshold=0.7)
rules
```

```python
rules.sort_values('lift',ascending = False)[0:30]
```

```python
rules[rules.lift>1]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.scatterplot(x=rules['support'], y=rules['confidence'])
plt.show()
```

```python
rules[['support','confidence','lift']].hist()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.scatterplot(x='support', y='confidence', data=rules, hue='antecedents')
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
plt.show()
```
