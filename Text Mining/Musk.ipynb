```python
import pandas as pd
import numpy as np
musk = pd.read_csv("Elon_musk.csv", encoding='latin1')
musk
```

```python
musk_1 = musk.drop(columns=['Unnamed: 0'])
musk_1
```

```python
tweets=[Text.strip() for Text in musk_1.Text]
tweets=[Text for Text in tweets if Text]
tweets[0:100]
```

```python
tweets_text=' '.join(tweets)
tweets_text
```

```python
from nltk.tokenize import TweetTokenizer
tknzr = TweetTokenizer(strip_handles=True)
tweets_tokens=tknzr.tokenize(tweets_text)
print(tweets_tokens)
```

```python
tweets_tokens_text=' '.join(tweets_tokens)
tweets_tokens_text
```

```python
import string
no_punc_text=tweets_tokens_text.translate(str.maketrans('','',string.punctuation))
no_punc_text
```

```python
import re
no_url_text=re.sub(r'http\S+', '', no_punc_text)
no_url_text
```

```python
from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt')
text_tokens=word_tokenize(no_url_text)
print(text_tokens)
```

```python
len(text_tokens)
```

```python
import nltk
from nltk.corpus import stopwords

# Download the stopwords data
nltk.download('stopwords')

# Access the English stopwords
my_stop_words = stopwords.words('english')
sw_list = ['\x92', 'rt', 'ye', 'yeah', 'haha', 'Yes', 'U0001F923', 'I']
my_stop_words.extend(sw_list)

no_stop_tokens=[word for word in text_tokens if not word in my_stop_words]
print(no_stop_tokens)
```

```python
lower_words=[Text.lower() for Text in no_stop_tokens]
print(lower_words[50:200])
```

```python
from nltk.stem import PorterStemmer
ps=PorterStemmer()
stemmed_tokens=[ps.stem(word) for word in lower_words]
print(stemmed_tokens[50:200])
```

```python
import spacy
nlp=spacy.load('en_core_web_sm')
doc=nlp(' '.join(lower_words))
print(doc)
```

```python
lemmas=[token.lemma_ for token in doc]
print(lemmas)
```

```python
clean_tweets=' '.join(lemmas)
clean_tweets
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
tweetscv=cv.fit_transform(lemmas)
print(cv.vocabulary_)
```

```python
feature_names = cv.get_feature_names_out()
print(feature_names[50:200])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()

X=cv.fit_transform(nouns_verbs)
sum_words=X.sum(axis=0)

words_freq=[(word,sum_words[0,idx]) for word,idx in cv.vocabulary_.items()]
words_freq=sorted(words_freq, key=lambda x: x[1], reverse=True)

wd_df=pd.DataFrame(words_freq)
wd_df.columns=['word','count']
wd_df[0:10]
```

```python
from nltk import tokenize
sentences=tokenize.sent_tokenize(' '.join(tweets))
sentences
```

```python
sent_df=pd.DataFrame(sentences,columns=['sentence'])
sent_df
```

```python
from textblob import TextBlob
def sentiment_analysis(ds):
    sentiment = TextBlob(ds["sentence"]).sentiment
    return pd.Series([sentiment.subjectivity, sentiment.polarity])
# Adding Subjectivity & Polarity
sent_df[["subjectivity", "polarity"]] = sent_df.apply(sentiment_analysis, axis=1)
sent_df
```

```python
def analysis(score):
    if score < 0:
        return "Negative"
    elif score == 0:
        return "Neutral"
    else:
        return "Positive"

sent_df["analysis"] = sent_df["polarity"].apply(analysis)
sent_df
```

```python
import seaborn as sns
plt.figure(figsize=(10,5))
sns.barplot(x='analysis',y='subjectivity',data=sent_df)
```
