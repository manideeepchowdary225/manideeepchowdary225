```python
# Rename the column
Delivery_1 = Delivery.rename({'Delivery Time': 'DT', 'Sorting Time': 'ST'}, axis=1)
Delivery_1.head(5)
```

```python
import statsmodels.formula.api as smf
model = smf.ols('DT~ST', Delivery_1).fit()
model.params
```

```python
model.summary()
```

```python
model.rsquared
```

```python
model_pred = model.predict()
model_pred
```

```python
rmse_val = np.sqrt(np.mean((np.array(Delivery_1.DT)-np.array(model_pred))**2))
rmse_val
```

Log Transformation

```python
model_log = smf.ols('DT~np.log(ST)', Delivery_1).fit()
model_log.summary()
```

```python
model_log.rsquared
```

```python
log_pred = model_log.predict(Delivery_1.ST)
log_pred.head(5)
```

```python
rmse_log = np.sqrt(np.mean((np.array(Delivery_1.DT)-np.array(log_pred))**2))
rmse_log
```

```python
plt.scatter(Delivery_1.ST, Delivery_1.DT, color='red')
plt.plot(Delivery_1.ST, log_pred, color='black') # black line is predicted data
plt.xlabel('Sorting Time')
plt.ylabel('Delivery Time')
```

**Exponential** **Transformation**

```python
# Using Exponential Transformation
model_exp = smf.ols('np.log(DT)~ST',Delivery_1).fit()
model_exp.summary()
```

```python
model_exp.rsquared
```

```python
model_pred = model_exp.predict()
model_pred
```

```python
exp_pred = np.exp(model_pred)
exp_pred
```

```python
rmse_exp = np.sqrt(np.mean((np.array(Delivery_1.DT)-np.array(exp_pred))**2))
rmse_exp
```

```python
plt.scatter(Delivery_1.ST, Delivery_1.DT, color='red')
plt.plot(Delivery_1.ST, exp_pred, color='black')
plt.xlabel('Sorting Time')
plt.ylabel('Delivery Time')
```

Using Log and both square

Quad Transformation

```python
Delivery_1['ST_sqr'] = Delivery_1.ST*Delivery_1.ST
Delivery_1.head(5)
```

```python
model_quad = smf.ols('np.log(DT)~ST+ST_sqr', Delivery_1).fit()
model_quad.summary()
```

```python
model_quad.rsquared
```

```python
model_pred_1 = model_quad.predict(Delivery_1)
model_pred_1
```

```python
quad_pred = np.exp(model_pred_1)
quad_pred
```

```python
# Shifiting the last column to the first column
final_Delivery = Delivery_1.copy()
first_column = final_Delivery.pop('ST')
final_Delivery.insert(0,'ST',first_column)
final_Delivery.head(5)
```

```python
rmse_quad = np.sqrt(np.mean((np.array(Delivery_1.DT)-np.array(quad_pred))**2))
rmse_quad
```

```python
plt.scatter(Delivery_1.ST, Delivery_1.DT, color='red')
plt.plot(Delivery_1.ST, quad_pred, color='black')
plt.xlabel('Sorting Time')
plt.ylabel('Delivery Time')
```
